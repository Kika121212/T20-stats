```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, HTML

# Load the IPL data into a DataFrame
df = pd.read_csv(r"C:\Users\Kika\Documents\Python\Cricket Lab\T20 staats.csv")

```

```python
striker_stats = df.groupby('Striker').agg(
    runs_scored=pd.NamedAgg(column='Batter Runs', aggfunc='sum'),
    balls_faced=pd.NamedAgg(column='Batter Balls', aggfunc='count'),
    innings=pd.NamedAgg(column='Match ID', aggfunc='nunique')
).reset_index()

striker_stats['Strike Rate'] = (striker_stats['runs_scored'] / striker_stats['balls_faced'] * 100)

# Ensure the lengths match by aligning the indices
dismissal_counts = df.groupby('Striker')['Dismissal Type'].count()
striker_stats['Average'] = (striker_stats['runs_scored'] / striker_stats['Striker'].map(dismissal_counts))

high_scores = df.groupby('Striker')['Batter Total Runs'].max()
striker_stats['Best'] = striker_stats['Striker'].map(high_scores)

fours_counts = df[df['Batter Runs'] == 4].groupby('Striker')['Batter Runs'].count()
striker_stats['4s'] = striker_stats['Striker'].map(fours_counts).fillna(0).astype(int)

Sixers_counts = df[df['Batter Runs'] == 6].groupby('Striker')['Batter Runs'].count()
striker_stats['6s'] = striker_stats['Striker'].map(Sixers_counts).fillna(0).astype(int)

Dot_balls = df[df['Batter Runs'] == 0].groupby('Striker')['Batter Runs'].count()
striker_stats = striker_stats.set_index('Striker')
striker_stats['Dot balls %'] = (Dot_balls / striker_stats['balls_faced'] * 100).fillna(0)
striker_stats = striker_stats.reset_index()



# Display the striker_stats DataFrame in a professional table format with enhanced designs and borders
styled_table = striker_stats.style.set_table_styles(
    [{'selector': 'thead th', 'props': [('background-color', '#4CAF50'), ('color', 'white'), ('font-weight', 'bold'), ('border', '1px solid black')]},
     {'selector': 'tbody tr:nth-child(odd)', 'props': [('background-color', '#f9f9f9'), ('color', 'Black'), ('border', '1px solid black')]},
     {'selector': 'tbody tr:nth-child(even)', 'props': [('background-color', '#f2f2f2'), ('color', 'Black'), ('border', '1px solid black')]}]
).set_properties(**{'text-align': 'Left', 'border': '1px solid black'}).hide(axis='index')

Batting_stats = display(HTML(styled_table.to_html()))


```
